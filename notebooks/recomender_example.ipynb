# Ejemplo de Sistema de Recomendación con CatBoost

Este notebook muestra cómo usar el sistema de recomendación implementado con CatBoost.

```python
import pandas as pd
import numpy as np
from src.model import RecommendationModel
from src.preprocessing import DataPreprocessor
from src.evaluation import Evaluator

# Crear un ejemplo de datos
np.random.seed(42)

# Generar datos de ejemplo
n_users = 100
n_items = 1000
n_interactions = 5000

# Crear IDs de usuarios y items
user_ids = [f'U{i}' for i in range(n_users)]
item_ids = [f'I{i}' for i in range(n_items)]

# Generar interacciones aleatorias
interactions = {
    'user_id': np.random.choice(user_ids, n_interactions),
    'item_id': np.random.choice(item_ids, n_interactions),
    'relevance': np.random.randint(1, 6, n_interactions)  # Relevancia de 1 a 5
}

# Convertir a DataFrame
interactions_df = pd.DataFrame(interactions)

# Mostrar las primeras filas
interactions_df.head()
```

```python
# Preprocesamiento de datos
preprocessor = DataPreprocessor()
processed_df = preprocessor.preprocess_interactions(interactions_df)

# Dividir en train y test
train_df, test_df, train_group_id, test_group_id = 
    preprocessor.prepare_training_data(processed_df)

# Mostrar tamaño de los conjuntos
print(f'Tamaño del conjunto de entrenamiento: {len(train_df)}')
print(f'Tamaño del conjunto de test: {len(test_df)}')
```

```python
# Inicializar y entrenar el modelo
model = RecommendationModel(learning_rate=0.1, iterations=1000)

# Preparar datos para el modelo
X_train, y_train = model.prepare_data(train_df)

# Entrenar el modelo
model.train(X_train, y_train, train_group_id)
```

```python
# Evaluar el modelo
evaluator = Evaluator()
ndcg_score = evaluator.evaluate_recommender(model, test_df)
print(f'NDCG Score: {ndcg_score:.4f}')
```

```python
# Hacer recomendaciones para un usuario específico
user_id = train_df['user_id'].iloc[0]  # Tomar el primer usuario del conjunto de entrenamiento

# Obtener todos los items disponibles como candidatos
candidates_df = pd.DataFrame({
    'user_id': [user_id] * n_items,
    'item_id': item_ids
})

# Obtener recomendaciones
recommendations = model.recommend(user_id, candidates_df)
print(f'Recomendaciones para el usuario {user_id}:')
recommendations
```
